```python
from tkinter import *
from tkinter import ttk
from ttkthemes import ThemedTk

window = ThemedTk(theme = 'arc')
window.title("CHD Predictor")
window.resizable(False, False)

#male or female
Sex = Label(window, text='Sex:')
Sex.grid(column=0, row=0)
m_f = ttk.Combobox(window)
m_f['values'] = ('Male','Female')
m_f.current(0)
m_f.grid(column=1, row=0)

#age 
age = Label(window, text="Age")
age.grid(column=0, row=1)
txt = Entry(window,width=10) 
txt.grid(column=1, row=1)

#BMI
BMI = Label(window, text='Current BMI value:')
BMI.grid(column=0, row =2)
BMI_txt = Entry(window, width=10)
BMI_txt.grid(column=1, row=2)

#Cigarettes per day 
cig = Label(window, text='Cigarettes per day:')
cig.grid(column=0, row =3)
cig_txt = Entry(window, width=10)
cig_txt.grid(column=1, row=3)


#Sys BP
sys_BP = Label(window, text='Systolic Blood pressure:')
sys_BP.grid(column=0, row =4)
sys_BP_txt = Entry(window, width=10)
sys_BP_txt.grid(column=1, row=4)


#Dia BP
dia_BP = Label(window, text='Diastolic Blood pressure:')
dia_BP.grid(column=0, row =5)
dia_BP_txt = Entry(window, width=10)
dia_BP_txt.grid(column=1, row=5)

#Heart Rate
HR = Label(window, text='Heart Rate:')
HR.grid(column=0, row =6)
HR_txt = Entry(window, width=10)
HR_txt.grid(column=1, row=6)

#Blood pressure medication 
BP_med = Label(window, text='Currently taking blood pressure medication')
BP_med.grid(column=0, row=7)
BP_btn = Checkbutton(window)
BP_btn.grid(column=1, row=7)


#Total Cholestrol 
chol = Label(window, text='Total Blood Cholestrol:')
chol.grid(column=0, row =8)
chol_txt = Entry(window, width=10)
chol_txt.grid(column=1, row=8)

#Prevalent Hypertension
prev_hyp = Label(window, text='Prevelant hypertension')
prev_hyp.grid(column=0, row=9)
prev_hyp = Checkbutton(window)
prev_hyp.grid(column=1, row=9)

#Prevalent Stroke
prev_str = Label(window, text='Prevelant stroke')
prev_str.grid(column=0, row=10)
prev_str = Checkbutton(window)
prev_str.grid(column=1, row=10)

#Diatbetes
Dia= Label(window, text='Currently type 2 diabetic ')
Dia.grid(column=0, row=11)
Dia_btn = Checkbutton(window)
Dia_btn.grid(column=1, row=11)


btn = Button(window, text="Run Predictor") 
btn.grid(column=4, row=12)
window.mainloop()
```
